# Importing libraries

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

# Reading the data

```python
df = pd.read_csv(r'C:\Users\Mateus de Alencar\Downloads\PortfolioProjects\Music\Spotify.csv', delimiter="|")
df
```

##### Deleting unnecessary columns

```python
df = df.drop(['index','added','bpm','nrgy','dnce','dB','live','val','dur','acous','spch','pop'],axis=1)
df
```

```python
df['top genre'].value_counts()
```

##### Creating another dataframe in order to manipulate the data

```python
df2 = df
```

##### Filling the missing values with "Unknown"

```python
df2.fillna(value="Unknown",inplace = True)
df2.head(30)
```

```python
df4 = df2
df4['genre'] = df2['top genre']
```

##### Replacing the various mixed genres into selected genres, for easier manipulation

```python
for i in df4['genre']:
    if 'pop' in i:
        df4.replace(i,'pop',inplace=True)
    elif 'hip hop' in i:
        df4.replace(i,'hip hop',inplace=True)
    elif 'indie' in i:
        df4.replace(i,'indie',inplace=True) 
    elif 'rock' in i:
        df4.replace(i,'rock',inplace=True)
    elif 'rap' in i:
        df4.replace(i,'rap',inplace=True) 
    elif 'house' in i:
        df4.replace(i,'house',inplace=True) 
    elif 'r&b' in i:
        df4.replace(i,'r&b',inplace=True) 
    elif 'funk' in i:
        df4.replace(i,'funk',inplace=True) 
    elif 'sertanejo' in i:
        df4.replace(i,'sertanejo',inplace=True) 
    elif 'reggaeton' in i:
        df4.replace(i,'reggaeton',inplace=True)
    elif 'adult standards' in i:
        df4.replace(i,'adult standards',inplace=True)
    else:
        df4.replace(i,'other',inplace=True)   
```
